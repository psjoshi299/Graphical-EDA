```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv(r"C:\Users\prathamesh\Downloads\medal.csv")
```

```python
df1
```

```python
country_data = df1["Country"]
medal_data = df1["Gold Medal"]
```

```python
plt.pie(medal_data, labels=country_data,autopct='%1.1f%%')
plt.title("Gold medal achievements of five most successful\n"+"countries in 2018 Summer Olympics")
plt.show()
```

```python
plt.bar(country_data,medal_data)
plt.show()
```

# Bar Chart - Plot Medal_data across each Country_data

```python
# plotting bar chart and setting bar width to 0.5 and aligning it to center
plt.bar(country_data,medal_data, width= 0.5, align='center', edgecolor='Orange',color='cyan')

# Adding and formatting title
plt.title("Gold Medal Across Countries\n", fontdict={'fontsize': 20, 'fontweight' : 'heavy', 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Countries", fontdict={'fontsize': 12, 'fontweight' : 'light', 'color' : 'Brown'})
plt.ylabel("Gold Medal", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})

plt.show()
```

```python
plt.hist(medal_data, bins = 10,edgecolor='Orange',color='cyan')

plt.show()
```
